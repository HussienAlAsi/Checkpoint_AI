```python
#Question2 
def factorial(x):
    if x ==1:
        return 1
    else:
        return x * factorial(x-1)
n = int(input('Enter Number :'))
print(factorial(n))
```

```python
#Question4
def missing_char(str, n):
    first_half = str[:n]
    second_half = str[n+1:]
    return first_half + second_half
missing_char('kitten',5)
```
